```python
from __future__ import division
import numpy as np
from matplotlib import pyplot as plt
import copy

import pyhsmm
import pyhsmm.internals.transitions as transitions
pyhsmm.internals.states.use_eigen() # makes HMMs faster, message passing done in C++ with Eigen
from pyhsmm.util.text import progprint_xrange
from itertools import product
```

### Get the data

```python
# Currently, synthetic
# Set parameters
N = 4
T = 1000
obs_dim = 10
obs_hypparams = {'mu_0':np.zeros(obs_dim),
                'sigma_0':np.eye(obs_dim),
                'kappa_0':0.1,
                'nu_0':obs_dim+5}
```

```python
# Setup the model
true_obs_distns = [pyhsmm.distributions.Gaussian(**obs_hypparams) for state in xrange(N)]
true_trans_distn = transitions.UniformTransitionsFixedSelfTrans(
        pi=pyhsmm.distributions.Multinomial(alpha_0=10.*N,K=N),
        lmbda=0.9)

truemodel = pyhsmm.models.HMM(
        init_state_concentration=6.,
        obs_distns=true_obs_distns,
        trans_distn=true_trans_distn)
```

```python
# Generate that data
data, labels = truemodel.generate(T)
```

```python
# Let's look at it for a sec, just a sec
figure(figsize=(20,8))
truemodel.plot()
# suptitle('True model')
```

### Define the model

```python
def fit_model(data, obs_hypparams, num_states, timescale_lambda, num_EM_attempts=5, verbose=False):
    # We import everything over again, because this code gets sent over to the IPython parallel engines
    import numpy as np
    from matplotlib import pyplot as plt
    import copy
    
    import pyhsmm
    import pyhsmm.internals.transitions as transitions
    pyhsmm.internals.states.use_eigen() # makes HMMs faster, message passing done in C++ with Eigen
    from pyhsmm.util.text import progprint_xrange
    from itertools import product
    
    if verbose:
        print 'Fitting with %d states, %2.2f lambda' % (num_states, timescale_lambda)
    
    # Build up the model
    # First the observation distribution
    obs_distns = [pyhsmm.distributions.Gaussian(**obs_hypparams) for state in xrange(num_states)]
    
    # Then the transition distribution (we have a special, semi-crippled version
    # that helps us figure out the right timescales, but doesn't care about the right interactions)
    trans_distn = transitions.UniformTransitionsFixedSelfTrans(
            pi=pyhsmm.distributions.Multinomial(alpha_0=10.*num_states,K=num_states),
            lmbda=timescale_lambda)
    
    # Now, instantiate the model
    fitmodel = pyhsmm.models.HMM(
            init_state_concentration=6.,
            obs_distns=obs_distns,
            trans_distn=trans_distn)

    # Add data to the model, which we'll fit over in a second
    fitmodel.add_data(data)
    
    # Alright, here's the meat.
    # We're going to do multiple passes of EM, because it can get stuck easily.
    theseBICs = []
    for i in range(num_EM_attempts):
        if verbose: print "\tEM Pass %d" % i
        # First, use Gibbs sampling to get a good start on the model
        if verbose: print "\t - Gibbs sampling"
        for itr in range(50):
            fitmodel.resample_model()
        
        # Then, do EM on the initialization.
        if verbose: print "\t - EM steps"
        for itr in range(50):
            fitmodel.EM_step()
        
        #Finally, save out the BIC of the model.
        theseBICs.append(fitmodel.BIC())
        
    return fitmodel, theseBICs
```

### Fit the model

```python
# Grid parameters
# TODO: lambda needs to be a list
all_timescale_lambda = [0.9]
all_num_states = range(2,10)
num_EM_attempts = 5
```

```python
# Non-parallel implementation
BICs = []
examplemodels = []

for num_states, timescale_lambda in product(all_num_states, all_timescale_lambda):
    fitmodel, theseBICs = fit_model(data, obs_hypparams, num_states, timescale_lambda)
    examplemodels.append(copy.deepcopy(fitmodel))
    BICs.append(theseBICs)
```

```python
# Parallel implementation
from IPython.parallel import Client
rc = Client()
dviews = rc[:]
states, lambdas = zip(*product(all_num_states, all_timescale_lambda))
n = len(states)
res = dviews.map_sync(fit_model, [data]*n, [obs_hypparams]*n, states, lambdas)

BICs = []
examplemodels = []
for fitmodel, theseBICs in res:
    BICs.append(theseBICs)
    examplemodels.append(copy.deepcopy(fitmodel))
```

```python
# Let's see how well it did!
plt.figure()
plt.errorbar(
        x=all_num_states,
        y=[np.mean(x) for x in BICs],
        yerr=[np.std(x) for x in BICs],
        )
plt.xlabel('num states')
plt.ylabel('BIC')

plt.figure()
examplemodels[np.argmin([np.min(x) for x in BICs])].plot()
plt.title('a decent model')

plt.show()
```
