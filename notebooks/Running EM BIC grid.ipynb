### Define the model

```python
def fit_model(data, obs_hypparams, num_states, timescale_lambda, num_EM_attempts=5, verbose=True):
    # We import everything over again, because this code gets sent over to the IPython parallel engines
    import numpy as np
    import copy
    import pyhsmm
    import pyhsmm.internals.transitions as transitions
    pyhsmm.internals.states.use_eigen() # makes HMMs faster, message passing done in C++ with Eigen
    from itertools import product
    
    if verbose:
        print 'Fitting with %d states, %2.2f lambda' % (num_states, timescale_lambda)
    
    # Build up the model
    # First the observation distribution
    obs_distns = [pyhsmm.distributions.Gaussian(**obs_hypparams) for state in xrange(num_states)]
    
    # Then the transition distribution (we have a special, semi-crippled version
    # that helps us figure out the right timescales, but doesn't care about the right interactions)
    trans_distn = transitions.UniformTransitionsFixedSelfTrans(
            pi=pyhsmm.distributions.Multinomial(alpha_0=10.*num_states,K=num_states),
            lmbda=timescale_lambda)
    
    # Now, instantiate the model
    fitmodel = pyhsmm.models.HMM(
            init_state_concentration=6.,
            obs_distns=obs_distns,
            trans_distn=trans_distn)

    # Add data to the model, which we'll fit over in a second
    fitmodel.add_data(data)
    
    # Alright, here's the meat.
    # We're going to do multiple passes of EM, because it can get stuck easily.
    theseBICs = []
    for i in range(num_EM_attempts):
        if verbose: print "\tEM Pass %d" % i
        # First, use Gibbs sampling to get a good start on the model
        if verbose: print "\t - Gibbs sampling"
        for itr in range(50):
            fitmodel.resample_model()
        
        # Then, do EM on the initialization.
        if verbose: print "\t - EM steps"
        for itr in range(50):
            fitmodel.EM_step()
        
        #Finally, save out the BIC of the model.
        theseBICs.append(fitmodel.BIC())
        
    return fitmodel, theseBICs
```

### Fit some models

```python
# Grid parameters
all_timescale_lambda = arange(0.1, 1.0, 0.1)
all_num_states = range(2,30)
num_EM_attempts = 5
```

```python
# THIS IS THE MEATY STUFF
# Parallel implementation
from IPython.parallel import Client
rc = Client()
dviews = rc[:]
states, lambdas = zip(*product(all_num_states, all_timescale_lambda))
n = len(states)
res = dviews.map_sync(fit_model, [data]*n, [obs_hypparams]*n, states, lambdas)

BICs = []
examplemodels = []
for fitmodel, theseBICs in res:
    BICs.append(theseBICs)
    examplemodels.append(copy.deepcopy(fitmodel))
```

### Show the grid scan of the model

```python
# Optionally, we might try to make a Hinton map to represent variance
from matplotlib import cm
def hinton(colors, sizes, colors_range=None, sizes_range=None, colormap=cm.jet, bgcolor='w'):
    """
    Draws a Hinton diagram for visualizing a weight matrix.
    """
    import matplotlib.pyplot as plt
    from matplotlib.patches import Rectangle, Circle
    from matplotlib.ticker import NullLocator

    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)

    sizes_range = [sizes.min(), sizes.max()] if sizes_range is None else sizes_range
    sizes = (sizes-sizes_range[0])/(sizes_range[1]-sizes_range[0])
    colors_range = [colors.min(), colors.max()] if colors_range is None else colors_range
    colors = (colors-colors_range[0])/(colors_range[1]-colors_range[0])


    ax.patch.set_facecolor(bgcolor)
    ax.set_aspect('equal', 'box')
    ax.xaxis.set_major_locator(NullLocator())
    ax.yaxis.set_major_locator(NullLocator())

    for (x,y),w in np.ndenumerate(sizes):
        color = colormap(colors[int(x),int(y)])
        size = np.abs(w)
       # patch = Rectangle([x - size / 2, y - size / 2], size, size, facecolor=color, edgecolor=color)
        patch = Circle([x - size / 2, y - size / 2], radius=size/2.0, facecolor=color, edgecolor=color)
        ax.add_patch(patch)
    ax.autoscale_view()

    # Reverse the yaxis limits
    ax.set_ylim(*ax.get_ylim()[::-1])
```
